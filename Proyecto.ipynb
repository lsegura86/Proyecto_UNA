```python
# Import packages
import pandas as pd
import re
```

```python
# Import data file to enviroment
data = pd.read_csv('Superstore.csv' , encoding = "ISO-8859-1", sep=';', decimal=',')
```

```python
# Explorer, Modify and Clean data
data['Month'] = data['Order Date'].str.split('/').str.get(1)
data['Year'] = data['Order Date'].str.split('/').str.get(2)
data.drop(columns=['Postal Code', 'Order ID', 'City', 'Quantity'])
```

```python
# Reporte Total de Ventas en la region sea igual a 
region = 'West'
ventasRegion = pd.DataFrame(data.loc[data['Region'] == region].groupby(['Region', 'State'])['Sales'].sum())
ventasRegion
```

```python
# Reporte Total de Ventas por anio y categoria
ventasAnioCate = pd.DataFrame(data.groupby(['Year', 'Category'])['Sales'].sum())
ventasAnioCate
```

```python
# Reporte Top 10 Ventas por Cliente
ventasClienteAnio = pd.DataFrame(data.groupby(['Customer Name'])['Sales'].sum())
ventasClienteAnio.sort_values(['Sales'], ascending=False).head(10)
```

```python
# Reporte Cantidad de clientes por Dominio de Correo

# Metodo para aplicar Exprecion Regular al correo y obtener dominio
def dominio_email(email):
    match = re.search(r'([\w\.-]+)@([\w\.-]+)', email)
    if match:
        return match.group(2)

# Se agrega la columna de Dominio
data['Dominio'] = data['Mail'].apply(dominio_email)

dominios = pd.DataFrame(data.groupby('Dominio')['Mail'].count())
dominios
```
